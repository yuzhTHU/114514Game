```python
from NGame.iiyokoiyo import load_model, load_env

for target in range(100+1):
    env = load_env()
    model = load_model('GP', env=env)
    result = model.solve(target=target, n_episode=1000, stop_threshold=0.0, quiet=True)
    print(f'{target} = {result}')
```

```python
from math import pi
from NGame.iiyokoiyo import load_model, load_env
from NGame.eqtree import *

env = load_env(symbols=[Add, Sub, Mul, Div, Cat, Sin])
model = load_model('GP', env=env)
result = model.solve(target=pi, n_episode=1000)
print(f'{pi} ~ {result} = {result.eval()}')
```
